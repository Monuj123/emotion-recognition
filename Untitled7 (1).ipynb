```python
from deepface import DeepFace
```

```python
img = cv2.imread('happy.jpeg')
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
prediction=DeepFace.analyze(img_rgb, enforce_detection=False)
```

```python
prediction
```

```python
type(prediction)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
print("Haar Cascade classifier loaded successfully.")
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print("Image converted to grayscale.")
```

```python
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
print(f"Detected {len(faces)} face(s).")
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
print("Rectangles drawn around detected faces.")
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Detected Faces')
plt.axis('off')
plt.show()
```

```python
dominant_emotion = prediction[0]['dominant_emotion']

# Define text properties
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_color = (0, 0, 255) # Red color in BGR
font_thickness = 1



# Add the text to the image
cv2.putText(img, dominant_emotion, (0,50), font, font_scale, font_color, font_thickness, cv2.LINE_AA)

# Display the image with the text
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Detected Faces with Dominant Emotion')
plt.axis('off')
plt.show()
```

```python

import cv2
from deepface import DeepFace

faceCascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

cap = cv2.VideoCapture(1)

# Check if the webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Analyze emotion
    result = DeepFace.analyze(frame, actions=['emotion'])

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    font = cv2.FONT_HERSHEY_SIMPLEX

    # Insert emotion text on video
    cv2.putText(
        frame,
        result['dominant_emotion'],
        (0, 50),
        font,
        3,
        (0, 0, 255),
        2,
        cv2.LINE_4
    )

    cv2.imshow('Original video', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
